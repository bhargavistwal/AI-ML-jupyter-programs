```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

data = {

    "Model" : np.random.choice(["swift","baleno","nexon","creta","i20","venue","city","verna","seltos","sonnet"],20),
    "Fuel_Type" : np.random.choice(["Petrol","Diesel"],20),
    "Price" : np.random.randint(400000,4600000,20),
    "Colour" : np.random.choice(["black","blue","grey","silver","white","red"],20),
    "Milage" : np.random.randint(12,25,20),
    "Top_Speed" : np.random.randint(140,250,20),
    "Engine_cc" : np.random.randint(800,3000,20),
    "Manufactoror" : np.random.choice(["maruti","hyundai","tata","honda","kia","toyota"],20)

}

df = pd.DataFrame(data)
print(df)
print("\n")
print(df.describe())

print("\n")
plt.bar(df["Model"],df["Price"])
plt.xlabel("Model")
plt.ylabel("Price")
plt.title("PRICES PER MODELS OF CARS")
plt.show()

print("\n")
plt.bar(df["Model"],df["Milage"])
plt.xlabel("Model")
plt.ylabel("Milage")
plt.title("MODELS AND THEIR MILAGE")
plt.show()

print("\n")
plt.bar(df["Manufactoror"],df["Engine_cc"])
plt.xlabel("Manufactoror")
plt.ylabel("Engine (CC)")
plt.title("CAR BRANDS AND THEIR ENGINE CAPACITY")
plt.show()
```
